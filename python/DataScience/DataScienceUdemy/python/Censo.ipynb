# Tansformação de Variaveis Categoricas com OneHotEncoder

## No algoritimo anterior, ele havia um certo Preconceito, pois ele iria considerar a raça por pesos ex: black: 2 White :1 , ele iria considerar o preto melhor que o branco, entao abordaremos de outra forma

```python
from sklearn.preprocessing import LabelEncoder, OneHotEncoder
from sklearn.compose import ColumnTransformer 

ct = ColumnTransformer([("Country", OneHotEncoder(), [1,3,5,6,7,8,9,13])],    remainder = 'passthrough')
previsores = ct.fit_transform(previsores).todense()
```

```python
pd.DataFrame(previsores)
```
