# Combinar Dados

```python
# criando dataset
lista1 = ["A","B","C","D"]
lista2 = ["1º","2º","3º","4º"]
df1 = pd.Series(lista1)
df2 = pd.Series(lista2)
```

```python
print(df1)
print("************************************")
print(df2)
```

```python
# APPEND (adicionará linhas de df2 ao final de df1)
df1.append(df2)
```

```python
# CONCAT (adicionará colunas de df2 a df1 se possuirem mesmo tamanho)
pd.concat([df1,df2], axis=1)
```

```python
# JOIN - DataFrames (unifica colunas em df1 com df2: left,right,outer, inner; precisa de valores identicos)
esq = pd.DataFrame({'A': ['A0', 'A1', 'A2', 'A3'],
                     'B': ['B0', 'B1', 'B2', 'B3'],
                     'key': ['K0', 'K1', 'K0', 'K1']}) 
esq
```

```python
dire = pd.DataFrame({'C': ['C0', 'C1'],
                      'D': ['D0', 'D1']},
                      index=['K0', 'K1'])
dire
```

```python
result = esq.join(dire, on='key')
result
```
